```python
import pandas as pd
import numpy as np
```

```python
fights = pd.read_csv("model.csv", index_col=0)
```

```python
fights = fights.drop('AName', axis=1)
```

```python
fights = fights.drop('BName', axis=1)
fights = fights.drop('AOdds', axis=1)
fights = fights.drop('BOdds', axis=1)
```

```python
fights.head()
```

```python
fights["target"] = (fights["Winner"] == 0).astype("int")
```

```python
#fights = fights.drop('Winner', axis=1)
X = fights.drop('target',axis=1)
y = fights['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 1000, random_state = 42)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)

```

```python

fights=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
fights

```
